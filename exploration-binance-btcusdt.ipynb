```python
import pandas as pd

history = pd.read_json("Binance_BTCUSDT_1d_1546300800000-1577836800000.json")
history.rename(columns={
    0: "dt_open", 1: "open", 2: "high", 3: "low", 4: "close", 5: "volume",
    6: "dt_close", 7: "quote_asset_volume", 8: "num_trades", 9: "base_asset_volume",
    10: "quote_asset_volume", 11: "-"
}, inplace=True)
history.set_index("dt_open", inplace=True)
history
```

## Hypothesis One: Volume will increase/decrease with movements in the price

To determine this, we want to examine the following features of the dataset:

Feature | Description
--------|-----------------
`close` | The closing price of the symbol
`volume` | The volume of trades for the symbol


```python
history.plot.line(y=["close", "volume"])
```

This will require some form of normalisation/scaling.

```python
history["close"] = (history["close"]-history["close"].mean())/history["close"].std()
history["volume"] = (history["volume"]-history["volume"].mean())/history["volume"].std()
history.plot.line(y=["close", "volume"])
```

```python
history['close'].corr(history['volume'])
```

This shows a statistically significant relationship between the volume and the price - with a `p` value below 0.01 - as expected:

 Variable | Value
---------|---------
  `r`    | 0.3887
  `N`    | 366
  `p`    | < 0.01

Whilst this demonstrates that there is a *real-time* relationship, this isn't indicative of a *predictive* relationship.

With our data retrieved at 10 minute intervals, we can shift the price value through 3 places, and determine whether the volume is indicative of price movement in the period - i.e day - prior to that movement.

```python
history["close"] = history["close"].shift(periods=1)
history
```

```python
history.plot.line(y=["close", "volume"])
```

```python
history['close'].corr(history['volume'])
```

We've essentially proved the obvious: that the price is correlated directly with the volume.

However, this only looks at the price itself, and doesn't reflect the *movement* of the price. To do this, we'd have to calculate the movement as a difference from the previous day - introducing a new feature. Providing us with a dataset like:

Closing Price
Market Volume
Movement

The Closing Price would not need to be considered as part of statistical model though: we're not trying to link price to volume, but movement to volume. price is merely used during feature engineering so as to generate the movement.
